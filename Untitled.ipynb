```python
from data_harvesting import Harvesting
import cnn
```

```python
# sample harvested data
hr = Harvesting()
df = hr.load_data()
print(len(df))
df.head()
```

```python
X, Y = hr.split_into_chunks(30,1,1)
X_train, X_test, Y_train, Y_test = Harvesting.data_split(X, Y)
```

### Define model

```python
input_shape=X_train[0].shape
output_shape=1
cnn_model = cnn.CNN(layers=[
            {
                'type': 'sepconv1D',
                'args': {
                     'filters': 32,
                     'kernel_size': 5,
                     'activation': 'relu',
                     'input_shape': input_shape
                }
             },
            {
                'type': 'maxpool1D',
                'args': {
                    'pool_size': 2
                }
            },
            {
                'type': 'conv1D',
                'args': {
                    'filters': 100,
                    'kernel_size': 3,
                    'activation': 'relu'
                }
            },
            {
                'type': 'maxpool1D',
                'args': {
                    'pool_size': 2
                }
            },
            {
                'type': 'dropout',
                'args': {
                    'ratio': 0.15
                }
            },
            {
                'type': 'flatten',
                'args': None
            },
            {
                'type': 'dense',
                'args': {
                    'output': 250
                }
            },
            {
                'type': 'dropout',
                'args': {
                    'ratio': 0.2
                }
            },
            {
                'type': 'activation',
                'args': {
                    'function': 'relu'
                }
            },
            {
                'type': 'dense',
                'args': {
                    'output': output_shape
                }
            },
            {
                'type': 'activation',
                'args': {
                    'function': 'linear'
                }
            },
        ])

cnn_model.build_model()
# model.compile_model()
```

```python
cnn_model.compile_model()
```

```python
history = cnn_model.fit_model(X_train, Y_train, epochs=800,batch_size=32)
```

```python
cnn_model.visualise_history()
```

```python
cnn_model.evaluate(X_test, Y_test)
```

### As we can see the above graph is not totally accurate but it is able to learn the trend i.e. the stock is gonna go up or down .
